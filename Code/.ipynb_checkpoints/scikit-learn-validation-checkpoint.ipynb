```python
# Stack Functions
def creating_test_train_dataframes(df_target,df_id, df_features, seed,
                                        ratio_test):
    '''
        Objective: To split of all three pandas dataset (df_target, df_id, df_features)
                   between training and testing datasets
                   
        Arguments: target dataset,id_dataset, features dataset, seed value, split ratio
                 
                   
        Returns  : 2 split datasets of training and testing each 
                   for target, id, features


    '''
    strt_clm = df_target
    xtrain, xtest, ytrain, ytest , id_train, id_test=train_test_split \
                    (df_features, df_target,df_id,test_size=ratio_test, \
                    stratify=strt_clm,random_state=seed)
    print ("No. of True Cases in training data set for" , ytrain.values.ravel().sum())
    print ("No. of True Cases in testing data set for",ytest.values.ravel().sum())
    
    print ("Ratio of True Cases in training data set: " , round(ytrain.values.ravel().sum()/len(ytrain.values.ravel()),2))
    print ("Ratio of True Cases in testing data set: ", round(ytest.values.ravel().sum()/len(ytest.values.ravel()),2))
    return xtrain, xtest, ytrain, ytest , id_train, id_test
```

```python
def trig_gs_rf(X_re_sampled, y_re_sampled, parameters):
    '''
        Objective: To trigger grid search pipeline with Random forest
        Arguments: Final Features and target datasets, Parameters of grid search, 
                   flag to find out which model is being run, target variable name('rr_18_campaign_flag')
                   
        Returns  : Fitted model object with best parameters obtained from gridsearch
        Output   : Print shape of each dataset created

    '''
    selectFeatures = sklearn.feature_selection.SelectKBest()
    clf = RandomForestClassifier(random_state=1000)

    steps = [('feature_selection', selectFeatures),
            ('random_forest', clf)]

    pipeline = sklearn.pipeline.Pipeline(steps)
    cv = sklearn.grid_search.GridSearchCV(pipeline,cv=3, \
                        param_grid=parameters, scoring='neg_log_loss',verbose=10)
    cv.fit(X_re_sampled, y_re_sampled.values.ravel())
    
    return cv
```
